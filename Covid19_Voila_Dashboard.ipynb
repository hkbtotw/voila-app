# World Map

```python
import folium
import numpy as np
# creating world map using Map class
world_map = folium.Map(location=[11,0], tiles="cartodbpositron", zoom_start=2, max_zoom = 6, min_zoom = 2)
# iterate over all the rows of confirmed_df to get the lat/long
for i in range(0,len(dfConfirmed)):
    folium.Circle(
        location=[dfConfirmed.iloc[i]['lat'], dfConfirmed.iloc[i]['long']],
        fill=True,
        radius=(int((np.log(dfConfirmed.iloc[i,-1]+1.00001)))+0.2)*50000,
        color='red',
        fill_color='indigo',
    ).add_to(world_map)
    
world_map
```
